```python
#1d
import pandas as pd

df = pd.read_csv('GlobalLandTemperaturesByState.csv')

#filtered for columns and resolved setting with copy warning
df2 = df[['dt','AverageTemperature','State']].copy()

#filter years >2000
df2['dt'] = pd.to_datetime(df2['dt'])
df2 = df2[df2['dt'].dt.year>2000]

#filter states
states = ['Wyoming', 'Nebraska', 'South Dakota']
df2 = df2[df2['State'].isin(states)]

#Group by date and calc mean temp
df3 = df2.groupby('dt')['AverageTemperature'].mean().reset_index()
df3.columns = ['Date', 'Average_Temperature']

#change date into integer from string
df3['date'] = df3['Date'].dt.year + df3['Date'].dt.month*100
df3

```

```python
#1e (model equation)
import numpy as np

def func(x,m,b,a,c):
    return a*np.cos(b*(x + m)) + c

x = df3['date']
y = df3['Average_Temperature']
a = (y.max() - y.min())/2
b = 10* np.pi
m = 0
c = 10


initial_guess = [m, b, a, c]
p0 = initial_guess                                                               
```

```python
#1f
import scipy.optimize as fit
parameters, cov_matrix = fit.curve_fit(func, x, y, p0)
print(parameters)
print(cov_matrix)
```

```python
#1g

import matplotlib.pyplot as plt
plt.plot(df3['Date'], df3['Average_Temperature'], label = 'Data', color = 'orange')
plt.plot(df3['Date'], func(x, *parameters),label = 'Fitted Curve', linestyle = '--', color = 'black')
plt.xlabel('Date')
plt.ylabel('Average Temperature')
plt.title('Avg Temperature Over Time')
plt.legend()
plt.show()
#func(x, *parameters)
```

```python
#1h
import numpy as np

m_err, b_err, a_err, c_err = np.sqrt(np.diag(cov_matrix))
print(f"m : {m:.2f} ± {m_err:.2f}")
print(f"b : {b:.2f} ± {b_err:.2f}")
print(f"a : {b:.2f} ± {a_err:.2f}")
print(f"c : {b:.2f} ± {c_err:.2f}") 
# equation : a*np.cos(b*(x + m)) + c
print(f"Final equation: y = {a:.2f}cos({b:.2f}*(x + {m:.2f})) + {c:.2f}")
```
