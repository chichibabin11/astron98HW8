```python
from astropy.table import Table
import pandas as pd

data = Table.read('global_CCl4_MM.dat', format = 'ascii')
selected_columns = ['CCl4ottoyr','CCl4ottomon','CCl4ottoGLm', 'CCl4ottoGLsd'] #combine year and month 
extracted = data[selected_columns]
df = extracted.to_pandas()
df.columns = ['year','month','global_mean_concentration', 'global_mean_concentration_sd']
df['date'] = pd.to_datetime(df[['year', 'month']].assign(day=1))

import matplotlib.pyplot as plt

plt.errorbar(df['date'], df['global_mean_concentration'], yerr = df['global_mean_concentration_sd'])
plt.xlabel('Date')
plt.ylabel('Global Mean Concentration')
plt.title('Global Mean Concentration Over Time')
plt.show()

#what
```

```python

df['date2'] = [((x- df['date'].iloc[0]).days) for x in df['date']]
```

```python
import scipy.optimize as fit
import numpy as np

def func(x, m, b):
    return m * x + b

df.dropna(inplace = True)

popt, pcov = fit.curve_fit(func, df['date2'], df['global_mean_concentration'])

# Extract parameters and their errors
m, b = popt
m_err, b_err = np.sqrt(np.diag(pcov))

print(m)

# Print parameters and their errors
print(f"Slope (m): {m:.2f} ± {m_err:.2f}")
print(f"Intercept (b): {b:.2f} ± {b_err:.2f}")

# Print final equation
print(f"Linear model: y = {m:.2f}x + {b:.2f}")
# Calculate predicted values
y_pred = func(df['date2'], m, b)

# Calculate residuals
residuals = df['global_mean_concentration'] - y_pred

# Calculate reduced chi-squared value
N = len(df)
M = 2  # Number of parameters in the model (slope and intercept)
chi_squared = np.sum((residuals / df['global_mean_concentration_sd'])**2)
red_chi_squared = chi_squared / (N - M)

# Print reduced chi-squared value
print(f"Reduced Chi-Squared: {red_chi_squared:.2f}")

#set a command where if nan replace with mean

plt.errorbar(df['date'], df['global_mean_concentration'], yerr = df['global_mean_concentration_sd'])
plt.errorbar(df['date'], func(df['date2'], *popt), yerr = df['global_mean_concentration_sd'])
plt.xlabel('Date')
plt.ylabel('Global Mean Concentration')
plt.title('Global Mean Concentration Over Time')
plt.show()
```
